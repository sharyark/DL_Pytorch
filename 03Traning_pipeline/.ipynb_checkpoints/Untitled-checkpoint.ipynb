```python
import torch as t
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(X_train)
x_test = scaler.transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder .transform(y_test)
```

```python
# now change thses array in pytorch array
x_train_tensor = t.from_numpy(x_train)
x_test_tensor = t.from_numpy(x_test)
y_train_tensor = t.from_numpy(y_train)
y_test_tensor = t.from_numpy(y_test)
```

```python
# defining the model
class MySimpleNN():
    def __init__(self,X):
        self.weight = t.rand(X.shape[1],1,dtype=t.float64,requires_grad=True)
        self.bias = t.zeros(1,dtype=t.float64,requires_grad=True)
    def forward(self,X):
        z = t.matmul(X,self.weight)+self.bias
        y_pred = t.sigmoid(z)
        return y_pred
    def loss(self,y_pred,y):
        # Clamp predictions to avoid log(0)
        epsilon = 1e-7
        y_pred = t.clamp(y_pred, epsilon, 1 - epsilon)

        # Calculate loss
        loss = -(y_train_tensor * t.log(y_pred) + (1 - y_train_tensor) * t.log(1 - y_pred)).mean()
        return loss
```

```python
# parameter
learning_rate = 0.1
epochs = 10
```

```python
# tarining pipeline
# 1) create model
model = MySimpleNN(x_train_tensor)
```

```python
# define loop
for epoch in range(epochs):
# 2) forward pass
    y_pred = model.forward(x_train_tensor)
# 3) caculate loss
    loss = model.loss(y_pred,y_train_tensor)
# 4) backward pass
    loss.backward()

# 5) parameter update
    # model.weight -= learning_rate * model.weight.grad
    # model.bias -= learning_rate * model.bias.grad
    with t.no_grad():
        model.weight -= learning_rate * model.weight.grad
        model.bias -= learning_rate * model.bias.grad

    model.weight.grad.zero_()
    model.bias.grad.zero_()
    print(loss)
```
